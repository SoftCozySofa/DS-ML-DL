## Box plot

```python
DF.boxplot(column='SalePrice', by='MS Zoning', figsize=(12, 12))
```

## Histogram

```python
DF['Survived'].hist(figsize=(12, 12), bins=3, normed=0)
```

## Violin plot

```python
plt.figure(figsize=(12, 12))
sns.violinplot(DF['SalePrice'])
sns.despine()
```

## Scatter plot

```python
plt.figure(figsize=(12, 12))
plt.title('Scatter Plot')
plt.scatter(DF['Survived'], DF['Pclass'])
```

## Scatter matrix

```python
colors = {'Grvl': 'green', 'Pave': 'red'}
scatter_matrix(DF,
               figsize=(20, 20),
               diagonal='kde',
               c=DF['Street'].replace(colors),
              );
```

## Pie chart

```python
fig, ax = plt.subplots(1)
DF['Survived'].value_counts().plot(kind='pie', labels=['', ''],
                                   ax=ax, autopct='%.2f%%', fontsize=15)
ax.set_title(u'Survived')
ax.axis('equal')
ax.legend(labels=[u'0', u'1'], framealpha=0)
ax.set_ylabel('')
plt.tight_layout()
```
